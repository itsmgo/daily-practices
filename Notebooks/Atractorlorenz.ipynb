```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
arxiu = open('/home/marti/Documents/C++/dades.txt', 'rb')

data=[] 

for row in arxiu: 
    data.append(row[4:-3])
    
ndt = []

for i in range(len(data)):
    fila = []
    for j in range(3):
        fila.append(round(float(str(data[i])[2:-1].split("  ")[j]), 3))
    ndt.append(fila)
    
ndt = np.array(ndt, dtype=float)

xs = []
ys = []
zs = []

for i in range(len(ndt)):
    xs.append(ndt[i][0])
    ys.append(ndt[i][1])
    zs.append(ndt[i][2])
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.axes(projection='3d')

ax.plot(xs, ys, zs, '-b')
```
